```python
# Function to convert 'avg views' to a numeric value
def convert_avg_views(value):

    value = str(value).strip()  # Ensure the value is a string and remove any extra spaces
    if 'M' in value:
        return float(value.replace('M', '')) * 1000  # Convert millions to numeric
    elif 'K' in value:
        return float(value.replace('K', ''))
    else:
      return float(value)

# Apply the conversion function to the 'avg views' column
Data['avg views'] = Data['avg views'].apply(convert_avg_views)
# Display the updated 'avg views' column
Data['avg views'].head()
```

```python
# Function to convert 'avg likes' to a numeric value
def convert_avg_likes(value):

    value = str(value).strip()  # Ensure the value is a string and remove any extra spaces
    if 'M' in value:
        return float(value.replace('M', '')) * 1000  # Convert millions to numeric
    elif 'K' in value:
        return float(value.replace('K', ''))
    else:
      return float(value)

# Apply the conversion function to the 'avg likes' column
Data['avg likes'] = Data['avg likes'].apply(convert_avg_likes)

# Display the updated 'avg likes' column
Data['avg likes'].head()
```

```python
# Function to convert 'avg comments' to a numeric value
def convert_avg_comments(value):

    value = str(value).strip()  # Ensure the value is a string and remove any extra spaces
    if 'M' in value:
        return float(value.replace('M', '')) * 1_000_000  # Convert millions to numeric
    elif 'K' in value:
        return float(value.replace('K', '')) * 1_000  # Convert thousands to numeric
    else:
            return float(value)  # Assume numeric if no suffix

# Apply the conversion function to the 'avg comments' column
Data['avg comments'] = Data['avg comments'].apply(convert_avg_comments)

# Display the updated 'avg comments' column
Data['avg comments'].head()
```

```python
# Function to convert 'subscribers' to a numeric value
def convert_subscribers(value):

    value = str(value).strip()  # Ensure the value is a string and remove any extra spaces
    if 'M' in value:
        return float(value.replace('M', ''))

# Apply the conversion function to the 'subscribers' column
Data['Subscribers_in millions'] = Data['Subscribers'].apply(convert_subscribers)

# Display the updated 'subscribers' column
Data['Subscribers_in millions'].head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the function to create the heatmap with top 10 rows
def bivariate_cc(data, cat_col1, cat_col2,top_n=10):
    # Select the top `n` rows
    data_top_n = data.head(top_n)

    # Create a contingency table using the filtered data
    contingency_table = pd.crosstab(data_top_n[cat_col1], data_top_n[cat_col2])

    # Plot the heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(contingency_table, annot=True, cmap='Blues', fmt='d', cbar=True)
    plt.title(f'Heatmap of {cat_col1} vs {cat_col2} ')
    plt.xlabel(cat_col2)
    plt.ylabel(cat_col1)
    plt.show()

# Example usage with your DataFrame `df`:
# Replace 'channel name' and 'Category' with the actual column names in your DataFrame
bivariate_cc(Data, cat_col1='Audience Country', cat_col2='Category')
```

```python
Algorithm_names = ['Multiple Linear Regression','Polynomial Regression','Lasso Regression',
        'Ridge Regression','RandomForest Regressor',"Support Vector Regressor",
         'Xgboost Regressor']

models = {'mlr':mlr, 'poly':poly, 'lasso':lasso, 'ridge':ridge, 'rf':rf, 'svr':svr, 'xgb':xgb}
```

```python
X = pd.concat([X_train, X_test], axis = 0)
y = pd.concat([y_train, y_test], axis = 0)
```

```python
X.shape,y.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score

# Initialize empty lists before the loop
trainRMSE = []
testRMSE = []
trainscore = []
testscore = []
fit = []
crossvalscore = []

# Loop through all models
for name, model in models.items():

    if name == 'poly':
        # Prepare polynomial features
        ptrain = X_train.iloc[:, 0:56]
        ptest = X_test.iloc[:, 0:56]

        # Transform training and test sets using PolynomialFeatures
        pftrain = polyfeat.fit_transform(ptrain)
        pftest = polyfeat.transform(ptest)

        # Ensure the model is fitted on polynomial features
        model.fit(pftrain, y_train)

        # RMSE and R2 score
        trainRMSE.append(round(np.sqrt(mean_squared_error(y_train, model.predict(pftrain))), 2))
        testRMSE.append(round(np.sqrt(mean_squared_error(y_test, model.predict(pftest))), 2))
        trainscore.append(round(r2_score(y_train, model.predict(pftrain)), 2))
        testscore.append(round(r2_score(y_test, model.predict(pftest)), 2))

        trscore = r2_score(y_train, model.predict(pftrain))
        tescore = r2_score(y_test, model.predict(pftest))

        # Bias-Variance Trade-off
        if trscore >= 0.60 and tescore >= 0.60:
            if trscore > tescore:
                if trscore - tescore >= 0.10:
                    fit.append("Overfit")
                else:
                    fit.append("Goodfit")
            else:
                fit.append("Nofit")
        elif trscore < 0.60 and tescore < 0.60:
            if abs(trscore) == 0 and abs(tescore) == 0:
                fit.append("Nofit")
            else:
                fit.append("Underfit")
        else:
            fit.append("Nofit")

        # Cross-validation
        scores = cross_val_score(model, polyfeat.transform(X.iloc[:, 0:56]), y, cv=3)
        crossvalscore.append(round(scores.mean(), 2))

    else:
        # Ensure the model is trained on the original features
        model.fit(X_train, y_train)

        # RMSE and R2 score
        trainRMSE.append(round(np.sqrt(mean_squared_error(y_train, model.predict(X_train))), 2))
        testRMSE.append(round(np.sqrt(mean_squared_error(y_test, model.predict(X_test))), 2))
        trainscore.append(round(r2_score(y_train, model.predict(X_train)), 2))
        testscore.append(round(r2_score(y_test, model.predict(X_test)), 2))

        trscore = r2_score(y_train, model.predict(X_train))
        tescore = r2_score(y_test, model.predict(X_test))

        # Bias-Variance Trade-off
        if trscore >= 0.60 and tescore >= 0.60:
            if trscore > tescore:
                if trscore - tescore >= 0.10:
                    fit.append("Overfit")
                else:
                    fit.append("Goodfit")
            else:
                fit.append("Nofit")
        elif trscore < 0.60 and tescore < 0.60:
            if abs(trscore) == 0 and abs(tescore) == 0:
                fit.append("Nofit")
            else:
                fit.append("Underfit")
        else:
            fit.append("Nofit")

        # Cross-validation score
        scores = cross_val_score(model, X, y, cv=3)
        crossvalscore.append(round(scores.mean(), 2))

# After the loop, you can create the DataFrame
model_results = pd.DataFrame({
    'Model': list(models.keys()),  # Use model names from the 'models' dictionary
    'TrainedModel': list(models.values()),  # Use the trained model instances
    'TrainRMSE': trainRMSE,
    'TestRMSE': testRMSE,
    'TrainScore': trainscore,
    'TestScore': testscore,
    'Fit': fit,
    'CrossValScore': crossvalscore
})

# Display the results
model_results
```

```python
# Complete Model Evaluation Report

display(pd.DataFrame({'Model':Algorithm_names, 'Trainedmodel':models.values(),
                      'TrainRMSE':trainRMSE, 'TestRMSE':testRMSE,
                      'Trainscore':trainscore, 'Testscore':testscore,
                      'Crossvalscore':crossvalscore, 'Fit':fit}))
```
